```python

def load_cic_ids2017_data(csv_files=None):
    """
    Load CIC-IDS2017 dataset - either actual files or sample data
    """
    if csv_files:
        print(" LOADING ACTUAL CIC-IDS2017 FILES...")

        # CIC-IDS2017 typically has multiple files for different days/attacks
        dfs = []
        for csv_file in csv_files[:3]:  # Load first 3 files to manage memory
            print(f"   Loading {os.path.basename(csv_file)}...")
            try:
                # CIC-IDS2017 files can be large, so we load with optimized settings
                df_part = pd.read_csv(csv_file, encoding='utf-8', low_memory=False)
                dfs.append(df_part)
                print(f"       Shape: {df_part.shape}, Columns: {len(df_part.columns)}")

                # Show label distribution if available
                if 'Label' in df_part.columns:
                    print(f"       Label distribution:")
                    print(df_part['Label'].value_counts().head())

            except Exception as e:
                print(f"       Error: {e}")
                try:
                    # Try with different encoding
                    df_part = pd.read_csv(csv_file, encoding='latin-1', low_memory=False)
                    dfs.append(df_part)
                    print(f"       Success with latin-1 encoding! Shape: {df_part.shape}")
                except Exception as e2:
                    print(f"       Failed with latin-1: {e2}")

        if dfs:
            df = pd.concat(dfs, ignore_index=True)
            print(f" Combined dataset shape: {df.shape}")
            return df
        else:
            print(" Could not load any CSV files, creating sample data...")
            return create_cic_ids2017_sample()
    else:
        print(" Using sample CIC-IDS2017 dataset")
        return create_cic_ids2017_sample()

# Load the data
print(" Loading CIC-IDS2017 dataset...")
df = load_cic_ids2017_data(csv_files)
```
